# 데이터 그룹핑

### groupby 함수

```python
#필요 라이브러리 로딩
import numpy as np
import pandas as pd
```

```python
# 샘플 데이터프레임 생성
df = pd.DataFrame({
    '학과' : ['수학', '화학', '수학', '화학', '수학'],
    '이름' : ['로버트', '엔드류', '유진', '제이슨', '제이크'],
    '학년' : [1, 2, 3, 2, 3],
    '학점' : [1.5, 2.7, 3.5, 1.9, 4.0]
})

display(df)
```

```python
#데이터 프레임에 대하여 학과를 기준으로 그룹화 실행
df_dept = df.groupby('학과')
```

```python
print(df_dept)
```

```python
#학과별 기술통계 출력
df_dept.describe()
```

```python
#학과별 원소의 개수 출력
df_dept.count()
```

```python
df_dept.std()
```

### 1단계 그룹핑

```python
# Serise에 대한 1단계 그룹핑
dept =df['학점'].groupby(df['학과'])
print(dept)
```

```python
#학과 평균 성적 출력
dept.mean()
```

```python
# 학과 성적 표준편차 출력
dept.std()
```

```python
#group별 크기 출력
group_size = dept.size()
print(group_size)
```

```python
math = dept.get_group('수학')
print(math)
```

##### 이처럼 'groupby' 객체의 그룹연산 메서드는 다양합니다. 자주 사용되는 메서드는 아래와 같습니다.

- mean, median, min, max, std : 그룹 데이터 평균, 중앙, 최소, 최대, 표준편차 값
- size, count : 그룹 데이터의 갯수
- describe : 위 집계연산과 1사분위 3사분위 값을 포함하여 데이터프레임으로 나타냄
- agg, aggregate, apply : 원하는 그룹연산 합수를 만들어 agg, aggregate, apply에 전달하여 적용할 수 있음

### 2단계 그룹핑

```python
# 2단계 그룹핑

dept = df.groupby([df['학과'], df['학년']]) #학과와 학년으로 그룹화 진행
dept
```

```python
#그룹별 정보 도출
dept.describe()
```

```python
#그룹 평균값 도출
display(dept.mean())
```

```python
# 더 많은 데이터로 실습 진행

# 새로운 데이터프레임 생성
df2 = pd.DataFrame({
    '학과' : ['화학', '수학', '화학', '수학'],
    '이름' : ['엔디', '제니', '엘리스', '멜리샤'],
    '학년' : [1, 2, 3, 2],
    '학점' : [3, 4.2, 3.1, 4.5]
})

#df1과 df2 연결

concat_df = pd.concat([df, df2], axis=0)
concat_df.reset_index(inplace=True, drop=True)
display(concat_df)

```

```python
#학과와 학년별로 그룹핑
dept_and_year = concat_df.groupby([concat_df['학과'], concat_df['학년']])
```

```python
#학과 & 학년별 정보 출력
dept_and_year.describe()
```

```python
#그룹별 평균 출력
dept_and_year.mean()
```

```python
# 2단계 그룹화 인덱스 확인
dept_and_year.mean().index
```

### apply 함수

```python
#학과별 우수 성적자 3명 출력 함수
def top3_dept_scorer(df):
    return df.sort_values(by='학점', ascending=False)[:3]
```

```python
#학과 그룹바이 객체 행성
dept = concat_df.groupby('학과')
```

```python
#apply와 top3_dept_scorer 함수
dept.apply(top3_dept_scorer)
```

```python
#학과 우수 성적자 2명 출력 함수
def top2_scorer(series):
    return series.sort_values(ascending=False)[:2]
```

```python
dept['학점'].apply(top2_scorer)
```

```python
pd.DataFrame(dept['학점'].apply(top2_scorer))
```

### lambda식 활용

```python
# 새로운 데이터 프레임 생성
df = pd.DataFrame({
    'key' : ['A', 'B', 'B', 'C', 'A', 'C', 'A', 'B', 'C', 'A'],
    'data' : [1, 2, 2, 1, 3, 8, 2, 5, 3, 6]
})
```

```python
df
```

```python
# key별로 그룹화를 진행 후 data 값의 크기 별로 오름차순 정렬
df.groupby('key').apply(lambda x: x.sort_values(by='data'))
```

```python
# 위에서 apply함수를 이용하여 진행한 학과별 성적 탑3 학생 출력을 almbda 식으로 진행
dept = concat_df.groupby('학과')
```

```python
dept.apply(lambda x: x.sort_values(by='학점', ascending=False)[:3])
```

### groupby와 재귀 함수

```python
for dept, group in concat_df.groupby(concat_df['학과']):
    print('학과: {}'.format(dept))
    display(group)
```

```python
#2차 그룹핑 재귀

for (dept, year), group in concat_df.groupby([concat_df['학과'], concat_df['학년']]):
    print('-' * 25)
    print('학과 :', dept)
    print('학과 :', year)
    display(group)
```
