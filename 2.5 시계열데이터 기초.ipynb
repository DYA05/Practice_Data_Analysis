### pd.date_range

```python
# date_range : 시작일과 기간을 인수로 설정
pd.date_range('2020-1-1', '2020-05-31')
```

```python
# 월별 데이터 출력
pd.date_range('2020-01-01', '2020-05-31', freq='M') # freq = 'M' : 월별 데이터로 생성 인수 (각 월말일 기준)
```

```python
pd.date_range('2020-01-01', '2020-05-31', freq='MS') # freq = 'M' : 월별 데이터로 생성 인수 (각 월첫날 기준)
```

##### 자주 사용하는 freq 인수
- D : 일별(default 값)
- W : 주별(일요일 기준)
- W-MON : 주별(월요일 기준)
- M : 월별(월의 마지막 날)
- MS : 월별(월의 첫째 날)
- B : 주말을 제외한 평일 인덱스 (i.e 토, 일 제외)

```python
# 종료일 명시 x, periods 명시
pd.date_range(start='2020-1-1', periods=45)
```

```python
pd.date_range(start='2020-1-1', periods=45, freq='MS')
```

### 시계열의 shift

```python
np.random.seed(1) # 고정된 난수만 출력하도록 시드 설정

arr = np. random.randn(12) # 기우시안 정규분포 확률을 따르는 난수 12개 생성

ts_idx =pd.date_range('2020-1-1', periods=12, freq='MS') # 해당 난수값의 날짜 인덱스 생성

# 날짜인덱스와 어레이를 이용하여 시리즈 생성

ts = pd.Series(arr, index= ts_idx)# 시리즈 생성
print(ts)
```

```python
# 1개월치 데이터 이동
ts.shift(1)
```

```python
# 4개월치 데이터 이동
ts.shift(4)
```

```python
# 각 월 첫주 일요일을 기준으로 데이터 쉬프트
ts.shift(1,freq="W")
```

### diff(n) 행  -n 기간전 행

```python
# 1개월치 차분 값 연산
ts.diff(1) # 차분값 = 현재달 - 1달전값
```

```python
# 연산방식
ts[1] - ts[0]
```

```python
# 3월치 차분값 연산
ts.diff(3) # 차분 값 = 현재달 - 3달전
```

```python
ts[3] - ts[0]
```

### resample

```python
np.random.seed(1) # 고정된 난수 출력을 위해 시드 설정

arr = np.random.randn(365) # 기우시안 분포를 따르는 365일치 난수 생성
time_idx = pd.date_range('2021-1-1', periods=365, freq='D') # 2021년 365일치 기간 생성

ts = pd.Series(arr, index=time_idx)
print(ts)
```

### down-sampling

```python
# 다운셈플링 예제
ts.resample('M').mean() # 일별 데이터를 월별로 변환
```

### up_sampling

```python
# 업샘플링 예제
arr = np.random.randn(10) # 새로운 10개의 난수 생성
time_idx = pd.date_range('2021-1-1', periods=10, freq='W') # 주별 데이터 10개 생성

ts = pd.Series(arr, index=time_idx)
print(ts)
```

```python
# forward filling 방식
ts.resample('D').ffill().head(21)
```

```python
# backward filling : 각 기간의 마지막 날을 참고하여 결측값 보간
ts.resample('D').bfill().head(21)
```

## 시계열 데이터 시각화

```python
# 필요 데이터 사용
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #그래프 상세 속성 설정
```

```python
np.random.seed(1) # 난수 설정
ts_data = np.random.randn(365,3) # 365x3 행렬의 난수 데이터 생성
ts_idx = pd.date_range(start='2021-01-01', periods=365, freq='D') #365개의 시계열 인덱스 생성

ts_df = pd.DataFrame(ts_data, index=ts_idx)
ts_df.head()
```

```python
ts_df.columns = ['X', 'Y', 'Z']
ts_df.head()
```

```python
ts_df.plot() # 단순하게 데이터프레임의 메소드인 plot을 호출하면 됨
```

```python
#누적합 함수를 이용하여 알아보기 쉽게 각 시리즈를 변경
ts_df = ts_df.cumsum()
ts_df.plot()
plt.title('Time Series Plot of Ramdom Numbers (Cumulative)')
plt.show()
```

```python
ts_df.plot(figsize=(10,3))
plt.title('Time Series Plot of Ramdom Numbers (Cumulative)')
plt.xlabel('Date')
plt.ylabel('Unit')
plt.show()
```

```python
# 년, 월, 일이 서로다른 칼럼에 있을때
data = {
    'year' : [2008, 2008, 2009, 2010],
    'month' : ['Jan', 'Feb', 'Mar', 'Apr'],
    'date' : [31, 28, 29, 30],
    'ales_amount' : [847, 678, 930, 445]
}

ice_df = pd.DataFrame(data)
display(ice_df)
```

```python
ice_df.info()
```

```python
ice_df.index
```

```python
#날짜 값을 갖고있는 year, month, date를 문자열로 변경
ice_df['year'] = ice_df['year'].astype(str)
ice_df['month'] = ice_df['month'].astype(str)
ice_df['date'] = ice_df['date'].astype(str)
```

```python
ice_df.info()
```

```python
# 날짜 문자열 생성
date_str = ice_df['year'].values + '-' +ice_df['month'] + '-' + ice_df['date'].values
date_str = np.array(date_str)
date_str
```

```python
# datetime index 생성
ice_idx = pd.to_datetime(date_str)
print(ice_idx)
```

```python
# 인덱스값 변경
ice_df.index = ice_idx
print(ice_df)
```

```python
#불필요한 컬럼 제거
ice_df.drop(['year', 'month', 'date'], axis=1, inplace=True)
display(ice_df)
```

```python
# 시도표(시계열 라인 그래프 생성)
ice_df.plot(figsize=(8,3))
plt.show()
```
