# 정적 시각화 실습 - matplotlib 활용

### pyplot의 plot

```python
# 필요 라이브러리 로딩
import matplotlib.pyplot as plt
import numpy as np

# 브라우저 내부(inline)에 바로 그려지도록 해주는 코드
%matplotlib inline
```

```python
# 정규분포에서 난수를 추출

np.random.seed(0) # 동일한 난수가 나오게끔 시드설정

x = np.arange(50) # 0부터 49까지 1씩 증가하는 정수 생성
y = np.random.randn(50) # 정규분포 난수 생성
print(y)
```

#### 기본 그래프 그리기
   1. figure를 만들고 --> **plt.figure()**
   2. figure내 그래프를 그리고 --> **plt.plot(x,y)**
   3. 그래프를 출력 --> **plt.show()**

```python
# 기본적인 그래프 그리는 방법 plot(x,y)
plt.figure()
plt.plot(x,y)
plt.show()
```

#### figure 크기 설정 & 축 제목 표기

```python
# 타이틀 x축, y축 레이블 생성

plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.xlabel('Period (t)') # X 레이블 표기
plt.ylabel('Percent change') # Y 레이블 표기
plt.title('Change of Interset Rate over Last 50 Days') # 그래프 타이틀 설정
# plt.savefig('interest_rate.png')
plt.show()
```

#### 그래프에 데이터 추가

```python
np.random.seed(2) #고정된 난수 설정
z = np.random.randn(50) # 정규분포에서 50개 추출
print(z)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y, label='Korea') # label을 이용하여 각 선이 어떤 데이터를 나타내는지 표기
plt.plot(x, z, label='United States')
plt.xlabel('Period (t)')
plt.ylabel('Percent change')
plt.title('Change of Interest Rates over Last 50 Days')
plt.legend() # legend를 이용하면 범례 표기 가능
#plt.tight_layout() # 그림 사이즈에 맞게 변경하여출력
# plt.savefig('interest_rate.png')
plt.show()
```

### 여러 그래프 그리기 1 : subplot
**matplotlib**의 **pyplot**이 제공하는 **subplot** 메서드는 여러개의 그래프를 하나의 그림에 나타나도록 합니다.

**subplot**은 그림의 위치와 순서에 관련한 3개의 값을 설정해줘야 합니다, 따라서 최종적으로 우리가 사용하고자 하는 **subplot**의 커멘드는 **plt.subplot(행, 열, 인덱스)** 의 모습입니다.

```python
a = np.linspace(0.0, 3.0, 31) # 0과 3사이에 있는 실수 31개 균등한 간격의로 생성
print(a)
```

```python
sin = np.sin(2 * np.pi * a)
print(sin)

decaying_sin = np.sin(2 * np.pi * a) * np.exp(-a)
print(decaying_sin)
```

```python
plt.subplot(2,1,1) #2개의 행과 1개의 열을 사용, 그리고 첫번째(위)에 그리겠다고 명시
plt.plot(a, decaying_sin, '-g.') # 라인 : - / 초록색 : g / 도트: .  으로 설정
plt.title('Sine Graph Approaching to Zero')

plt.subplot(2, 1, 2) #2개의 행과 1개의 열을 사용, 그리고 두번째(아래)에 그리겠다고 명시
plt.plot(a, sin, 'o-') # 도트 : o / 라인: -  으로 설정
plt.title('Sine Graph')

plt.tight_layout()
# plt.savefig('sine_graphs.png') #그래프를 저장하는 커맨드
plt.show()
```

#### **sharex** : x축 공유

```python
plt.subplot(2,1,1) #2개의 행과 1개의 열을 사용, 그리고 첫번째(위)에 그리겠다고 명시
plt.plot(a, decaying_sin, '-g.', label='decaying_sin') # 라인 : - / 초록색 : g / 도트: .  으로 설정
plt.xticks(visible=False) #visible = False 설정
plt.title('Sine Graph Approaching to Zero')
plt.legend()

plt.subplot(2, 1, 2) #2개의 행과 1개의 열을 사용, 그리고 두번째(아래)에 그리겠다고 명시
plt.plot(a, sin, 'o-', label='sine') # 도트 : o / 라인: -  으로 설정
plt.title('Sine Graph')
plt.xlabel('Num Cycle')
plt.legend()

plt.suptitle('Two Different Sine Graph', y=1.05, fontsize=20) # suptiltle을 이용하여 타이틀 설정

plt.show()
```

### 여러 그래프 그리기 2 : subplots

```python
fig, ax = plt.subplots(2,1) # figure와 ax를 같이 설정 해줘야함
ax[0].plot(a, decaying_sin, 'g-') #ax[행 번호, 열 번호], 지금은 2개의 그래프이기 때문에 인덱스 하나로 표현 가능
ax[1].plot(a, sin, 'b-')
plt.show()
```

**subplot**과 **subplots**의 차이는 **state_based**와 **object-oriented**방식의 차이입니다.

### 두 종류의 그래프 하나에 담기 : 바 그래프 & 라인 그래프
아래의 데이터는 임의로 만든 **ABC**라는 가상의 국가의 1951년 부터 1960년까지 집 거래량과 연 평균 집값입니다.
   - 연 평균 집값 (housing_prices)
   - 총 거래량 (num_trading)
   - 기간 (periods)

```python
# 1. 토이데이터 준비

## ABC 국가의 연간 평균 집값과 총 거래량

periods = np.arange(1951, 1961) # 연도 표기 (1951년부터 1960년까지 생성)
housing_prices = np.array([15,25,31,24,22,27,40,55,60,58]) # unit = $'000
num_trading = np.array([6, 6.6, 7, 4.5, 4, 6, 9.5, 11, 20, 18]) #unit 백만
```

아래의 그래프는 **plt.subplot**을 활용하여 그려보겠습니다.
**plt.subplots**을 사용하려면 두개의 변수가 필요합니다.
**figure**를 나타낼 **fig** 변수와 **Axes** 객체를 나타낼 **ax**변수를 사용할 것입니다.
아래에서 사용될 **Axes** 객체의 함수 **twinx**에 대해 간략하게 짚고 넘어가겠습니다.

   - **twinx** : **Axes** 객체의 **twinx** 매서드는 특정 **ax**와 x축을 공유하는 새로운 **Axes**를 만들때 사용합니다.
   - ex) **ax2 = ax1.twinx()

위 함수를 이용하여 같은 **Axes**에 각기 다른 두개의 그래프를 그려보겠습니다.
   
   

```python
fig, ax1 = plt.subplots(figsize=(9,5)) # figsize 인수를 주어 figure의 크기를 정의
# ax1
# 집값 그래프는 라인으로 표기 plot 이용
ax1.plot(periods, housing_prices, 'o-', linewidth=1, alpha=0.8, label='housing price') # alpha는 투명도
ax1.set_xlabel('Year', fontsize=13)
ax1.set_ylabel('Price ($,000)', fontsize=13)
ax1.set_ylim(0,70) # ax1 의 y축 범위 설정 0~70까지 표기

#총 거래량 그래프는 막대그래프로 표기 bar 이용
ax2 = ax1.twinx() #twinx는 x축은 공유하지만 y축은 공유하지 않게 함
ax2.bar(periods, num_trading, color='coral', alpha=0.3, label='#houses traded') # bar생성
ax2.set_ylabel('#house traded (milions)', fontsize=13)
ax2.set_ylim(0, 22)

ax1.legend(loc='upper left') # legend에 loc 인수를 설정하면 범례의 위치를 설정할 수 있음
ax2.legend(loc='upper right')

plt.title('Housing Price and Demands of ABC Country between 1951 and 1960', fontsize=16)
plt.tight_layout()

plt.show()
```
