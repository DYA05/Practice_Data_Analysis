# DataFrame Merge

```python
#판다스 패키지 로딩
import pandas as pd
```

## inner join

```python
# (데이터 merage 실습용) 두개의 데이터 프레임 형성

data1 = {
    '학번' : [1, 2, 3, 4],
    '이름' : ['아이유', '김연아', '홍길동', '강감찬'],
    '학과' : ['철학', '경영학', '컴퓨터', '물리학']
}

data2 = {
    '학번' : [1, 2, 3, 5],
    '학년' : [2, 4, 3, 1],
    '학점' : [1.5, 2.0, 4.1, 3.8]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
# inner join -- 일치되는 것만 merge

inner_df = pd.merge(df1, df2, on="학번", how='inner')
```

```python
display(inner_df)
```

## outer join

```python
# outer join
outer_df = pd.merge(df1, df2, on='학번', how='outer')
```

```python
display(outer_df)
```

## left and right join

```python
left_df = pd.merge(df1, df2, on='학번', how='left')
display(left_df)
```

```python
right_df = pd.merge(df1, df2, on='학번', how='right')
display(right_df)
```

# merge 응용 : 현실 데이터를 다룰 때 생길만한 경우

   ## 1) 같은 의미 다른 컬럼명을 병합할때
   

```python
#case 1: 같은 의미 다른 컬럼명 ---> 학번 vs 학생고유번호
data1 = {
    '학번' : [1, 2, 3, 4],
    '이름' : ['아이유', '김연아', '홍길동', '강감찬'],
    '학과' : ['철학', '경영학', '컴퓨터', '물리학']
}

data2 = {
    '학생고유번호' : [1, 2, 3, 5],
    '학년' : [2, 4, 3, 1],
    '학점' : [1.5, 2.0, 4.1, 3.8]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
merged_df = pd.merge(df1, df2, left_on='학번', right_on='학생고유번호', how='inner')

display(merged_df)
```

```python
# 두개의 공통 키 중 하나를 삭제

#merged_df.drop('학생고유번호',axis=1)
merged_df.drop('학생고유번호', axis=1, inplace=True)
display(merged_df)
```

## 2) 공유하는 컬럼이 없는경우

```python
#왼쪽데이터
data1 = {
    '학번' : [1, 2, 3, 4],
    '이름' : ['아이유', '김연아', '홍길동', '강감찬'],
    '학과' : ['철학', '경영학', '컴퓨터', '물리학']
}

data2 = {
    '학년' : [2, 4, 3, 1],
    '학점' : [1.5, 2.0, 4.1, 3.8]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2, index=[1, 2, 4, 5]) # 예제를 위해 이상한 인덱스를 부여

display(df1)
display(df2)
```

```python
#양 데이터프레임의 인덱스로 병합 진행

merge_df1 = pd.merge(df1, df2, left_index=True, right_index=True, how='inner')

display(merge_df1)
```

```python
# 왼쪽은 학번기준으로 오른쪽은 인덱스를 기준으로 병합
merged_df2 = pd.merge(df1, df2, left_on='학번', right_index=True, how='inner')
display(merged_df2)
```

```python
# join 사용 : 양쪽 데이터프레임의 인덱스를 기준으로 병합
df1.join(df2, how='inner') # index기준
```

## 데이터 연결 pandas.concat

```python
#필요라이브러리 로딩
import numpy as np
import pandas as pd

np.random.seed(1) # 예제를 위한 고정된 난수설정
```

```python
# 3x2 정수로 구성된 메트릭스 생성
df1 = pd.DataFrame(np.random.randint(0, 9,(3,2)),
                  index = ['a', 'b', 'c'],
                   columns=['one', 'two']
                  )
                    
display(df1)
```

```python
df2 = pd.DataFrame(np.random.randint(0, 9,(2,2)),
                  index = ['a', 'b'],
                   columns=['three', 'four']
                  )
                    
display(df2)
```

```python
# df1 & df2 concat으로 연결
result_df = pd.concat([df1, df2], axis=1)
display(result_df)
```

```python
# outer_join으로도 concat과 같은 결과 데이터플레임을 확인할 수 있음
merge_df = pd.merge(df1, df2, left_index=True, right_index=True, how='outer')
display(merge_df)
```

```python
# 행방향 연결
result_df = pd.concat([df1, df2], axis=0)
display(result_df)
```
